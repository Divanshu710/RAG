```python
## create a simple txt file

import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_text = {
    "../data/text_files/python_intro.txt":"""Python Programming Introduction
Python is a high-level, interpreted programming language known for its simplicity and readability.
It supports multiple programming paradigms, including procedural, object-oriented, and functional programming.
Python's extensive standard library and active community make it a popular choice for web development, data analysis, artificial intelligence, scientific computing, and more.

Key Features:
- Easy to learn and use
- Interpreted language with dynamic typing
- Extensive standard library
- Cross-platform compatibility
"""
}

for file_path, content in sample_text.items():
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content)

print("Sample text files created!")
```

```python
sample_text["../data/text_files/machine_learning.txt"] = """Machine Learning Introduction
Machine learning is a subfield of artificial intelligence that enables computers to learn from data and improve over time.
Common techniques include regression, classification, clustering, and neural networks.
Applications include recommendation systems, fraud detection, and language translation.
"""

for file_path, content in sample_text.items():
    with open(file_path, "w", encoding="utf-8") as file_handle:
        file_handle.write(content)

print("Sample text files created/updated!")
```

```python
### TextLoader

# from langchain_core.document_loaders import TextLoader

from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")
doxument = loader.load()
print(doxument)
```

```python
### Directory Loader

from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader("../data/text_files", 
                             glob="**/*.txt",
                            loader_cls=TextLoader,
                            loader_kwargs={"encoding": "utf-8"},
                            show_progress=True
                            
                            )

documents = dir_loader.load()
print(documents)
```
